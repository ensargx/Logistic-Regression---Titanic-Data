```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
```

```python
data_test = pd.read_csv('/kaggle/input/titanic/test.csv')
data_train = pd.read_csv('/kaggle/input/titanic/train.csv')
```

```python
x_train = data_train.drop("Survived",axis=1)
y_train = data_train["Survived"]
x_test = data_test
```

```python
mean_age = x_train['Age'].mean()
x_train['Age'].fillna(mean_age, inplace=True)
```

```python
def ayarla(x):
    x = x.drop(['PassengerId','Name','Ticket','Cabin'],axis=1)
    x = pd.get_dummies(x, columns=['Sex', 'Pclass','Embarked'])
    return x
```

```python
x_train = ayarla(x_train)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(x_train)
```

```python
model = LogisticRegression(solver='saga',max_iter=10000)
model.fit(X_train_scaled,y_train)
```

```python
mean_age = x_test['Age'].mean()
x_test['Age'].fillna(mean_age, inplace=True)
mean_fare = x_test['Fare'].mean()
x_test['Fare'].fillna(mean_fare, inplace=True)
x_test = ayarla(x_test)
X_test_scaled = scaler.fit_transform(x_test)
```

```python
y_pred = model.predict(X_test_scaled)
```

```python
y_train_pred = model.predict(X_train_scaled)
```
